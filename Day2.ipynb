```python
import os
from dotenv import load_dotenv
from langchain.chat_models import ChatOpenAI
from langchain.prompts.few_shot import FewShotChatMessagePromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler, get_openai_callback
from langchain.prompts import ChatMessagePromptTemplate, ChatPromptTemplate
from langchain.globals import set_debug


load_dotenv()

chat = ChatOpenAI(
    temperature=0.1,
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ]
)

examples = [
    {
        "movie": "명량",
        "answer": """
        해당 영화에 대한 정보는 다음과 같습니다:
        감독: 김한민
        주요 출연진: 치민식,류승룡,조진웅
        예산 및 흥행수익 : 제작비 약 180억 원, 국내 관객 수 17,616,371명, 매출액 약 1,357억 원
        장르: 역사, 전쟁
        """,
    },
    {
        "movie": "극한직업",
        "answer": """
        해당 영화에 대한 정보는 다음과 같습니다:
        감독: 이병헌
        주요 출연진: 류승룡, 이하늬, 진선규, 이동휘, 공명
        예산 및 흥행수익 : 제작비 약 65억 원, 국내 관객 수 1,626만 명으로 역대 한국 영화 흥행 2위 기록
        장르: 코미디, 범죄
        """,
    },
    {
        "movie": "신과함께: 죄와 벌",
        "answer": """
        해당 영화에 대한 정보는 다음과 같습니다:
        감독: 김용화
        주요 출연진: 하정우, 차태현, 주지훈, 김향기
        예산 및 흥행수익 : 제작비 약 200억 원, 국내 관객 수 1,441만 명으로 역대 한국 영화 흥행 3위 기록
        장르: 판타지, 드라마
        """,
    },
]

example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "What do you know about {movie}?"),
        ("ai", "{answer}"),
    ]
)

example_prompt = FewShotChatMessagePromptTemplate(
    example_prompt = example_prompt,
    examples = examples
    
)

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a movie expert. Follow the given prompts to provide concise answers to the format."),
        example_prompt,
        ("human", "What do you know about {movie}?"),
    ]
)

chain = final_prompt | chat


final_prompt.format(movie="태극기 휘날리며")

with get_openai_callback() as usage:
    result = chain.invoke({"movie": "태극기 휘날리며"})
    print(f"{result}\n\n")
    print(usage)
```
